```
class MultiColumnLabelEncoder:
    def __init__(self,columns = None):
        self.columns = columns # array of column names to encode

    def fit(self,X,y=None):
        return self # not relevant here

    def transform(self,X):
        '''
        Transforms columns of X specified in self.columns using
        LabelEncoder(). If no columns specified, transforms all
        columns in X.
        '''
        output = X.copy()
        if self.columns is not None:
            for col in self.columns:
                output[col] = LabelEncoder().fit_transform(output[col])
        else:
            for colname,col in output.iteritems():
                output[colname] = LabelEncoder().fit_transform(col)
        return output

    def fit_transform(self,X,y=None):
        return self.fit(X,y).transform(X)
```

## Reduccio de dimensionalitat

### PCA

```
from sklearn import decomposition
from sklearn.preprocessing import StandardScaler
XS = StandardScaler().fit_transform(data)
```

```
XS.shape
```

```
#Aplicar PCA a los datos escalados
pca = decomposition.PCA(n_components=72).fit(XS)
#Proyección de los datos en el espacio PCA
Xproj = pca.transform(XS)
Xproj.shape
```

```
print('Valores propios = {} {} {} {} {} {} {} {} {} {} {} {} {} {}'.format(*pca.explained_variance_))
print('Fracción de la varianza explicada = {} {} {} {} {} {} {} {} {} {} {} {} {} {}'.format(*pca.explained_variance_ratio_))
```

```
100*pca.explained_variance_ratio_.cumsum()
```

```
# Reducir dimensionalidad de los datos quedandonos con 48 componentes PCA:
ndim_reducida = 48 # establecer dimensionalidad reducida a partir de la varianza explicada por cada componente
XPCA = Xproj[:,0:ndim_reducida]

print('Dimensionalidad datos en espacio PCA reducido = {}'.format(XPCA.shape))

data_reduced = pd.DataFrame(XPCA)
```

**ML con reduccion de dimensionalidad PCA**

```
def check_location (x,y):
  url = "https://nominatim.openstreetmap.org/reverse?lat={lat}&lon={lon}&format=geojson".format(lat = x, lon = y)
  t = requests.get(url)
  t.raise_for_status()
  datajson = t.json()
  
  features_key = 'features' in datajson
  if features_key:
    key1_present = 'house_number' in datajson['features'][0]['properties']['address']
    key2_present = 'city' in datajson['features'][0]['properties']['address']
    if key1_present and key2_present:
      house_number = datajson['features'][0]['properties']['address']['house_number']
      road = datajson['features'][0]['properties']['address']['road']
      city = datajson['features'][0]['properties']['address']['city']
      state = datajson['features'][0]['properties']['address']['state']
      postcode = datajson['features'][0]['properties']['address']['postcode']
      country = datajson['features'][0]['properties']['address']['country']
      full_address = (pd.Series([[house_number, road, city, state, postcode, country]])).str.join(' ')
      full_address_str = full_address[0]
    else:
      city = 'house_or_city_not_found'
      full_address_str = 'not found'
  else:
    city = 'features not present'
    full_address_str = 'not found'
  return city, full_address_str
```
